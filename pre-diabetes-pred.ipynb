```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data1 = pd.read_csv("../input/early-stage-diabetes-risk-prediction-dataset/diabetes_data_upload.csv")
data1.head()
```

```python
plt.hist(data1["Gender"])
data1["Gender"].value_counts()

from sklearn.preprocessing import OneHotEncoder
oh = OneHotEncoder()
oe_results = oh.fit_transform(data1[["Gender"]])
#pd.DataFrame(oe_results.toarray(), columns=oh.categories_)
data1 = pd.DataFrame(oe_results.toarray(), columns=oh.categories_).join(data1)
```

```python
cols=data1.columns.values
cols
```

```python
data1
```

```python
for col in cols:
    data1[col].replace(('Yes', 'No'), (1, 0), inplace=True)
data1.head()
```

```python
data1["Gender"].replace(('Male', 'Female'), (1, 0), inplace=True)
data1["class"].replace(('Positive', 'Negative'), (1, 0), inplace=True)
data1["Age"]=data1["Age"]/100
data1.head()
```

```python
data1 = data1.drop('Gender', 1)
data1
```

```python
data1.shape
```

```python
traindata=np.array(data1)
traindata.shape
trainX=traindata[:,0:17]
trainY=traindata[:,17]
'''from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
encoder = LabelEncoder()
encoder.fit(trainY)
encoded_Y = encoder.transform(trainY)
encoded_trainY = np_utils.to_categorical(encoded_Y)
print(encoded_trainY)'''
```

```python
import keras
import tensorflow as tf
input1 =tf.keras.layers.Input(15)
l1 = tf.keras.layers.Dense(50)(input1)
#l2 = tf.keras.layers.Dense(30)(l1)
output = tf.keras.layers.Dense(1, activation= "softmax")(l1)
model = tf.keras.Model(input1,output)
model.compile(loss='binary_crossentropy', optimizer= "adam",metrics=['accuracy'])
model.summary()
```

```python
model.fit(trainX,trainY, epochs=25, batch_size = 1, validation_split=.15)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
import pickle
logres1 = LogisticRegression(random_state=0).fit(trainX,trainY)
pred4 = logres1.predict(trainX)
#pickle.dump(logres1, open('./dia_pred.pkl','wb'))
print(metrics.classification_report(trainY, pred4))
print("f1 score: ", metrics.f1_score(trainY, pred4))
print("accu score: ", metrics.accuracy_score(trainY, pred4))
```
